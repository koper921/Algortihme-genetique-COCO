```python
def carre(x):
    return x*x



def func(x):
    if len(x)==1:
        return x**2
    else :
        s=0
        for i in range(len(x)):
            s=s+x[i]**2
        return s
    
def func1(x):
    if len(x)==1:
        return x**3
    else :
        s=0
        for i in range(len(x)):
            s=s+x[i]**3
        return s

# FONCTION POUR INTEGRER TON ALGO A COCO
def fitness( fun, *args ):
    return fun( *args )



def mutate(individual):
    
    element_muter =[]
    for element in indivudal:
        element_muter.append(element*uniform(-0.1, 0.1))
        
    return element_muter

# Given two parents, it returns the offspring of them
def crossover(male, female, n ):
    mx,my, _, _ = male
    _,_, fx, fy = female
    l= []
    l.append(mx)
    l.append(my)
    l.append(fx)
    l.append(fy)
    return l

# adapter aux variable appartenant à Rn
def crossover1(male, female, n ):
    l=[]
    
    for i in range(len(male)):
        if i < int(n/2):
            l.append(male[i])
    
    for i in range(len(female)):
        if i >= int(n/2):
            l.append(female[i])
    
    return l





    



def racine(x):
    return np.power(x,1/2)

def select_mu_best(individus, mu, n, func):
    result= []
    for element in individus :
        
        
        r = fitness(func, list(element))
        result.append(r)
        
    # les n plus faible selon f_ranking 
    index_min = heapq.nsmallest(mu, range(len(result)), key=result.__getitem__)

    parent= []
    for i in index_min:
        parent.append(individus[i])
    
    #fais un crossover des meilleurs parents 
    progeniture=[]
    for i in range(len(parent)):
            male, femelle= randint(0,mu-1), randint(0,mu-1)
            #sexes = np.random.randint(mu- 1,size= n)
            
            progeniture.append(crossover1(parent[male], parent[femelle], n))

    
    
    return progeniture

        
def norm_2(x):
    s =0
    for el in x :
        s= s + el**2
    return np.power(s, 1/2)

def norm_1(x):
    s=0
    for el in x :
        s = s + np.abs(el)
    return s
        

# la partie genetique fonctionne car elle fait bien baisser les valeurs,
# mais ce sont les valeurs de sigma qui font exploser les valeurs de x , ca ne converge pas ..
    

def algo_genetic( func,  x  , lambda_= 100,epochs = 1000):
    
    n=len(x)
    mu = int(lambda_/4)
    c_sigma =racine(mu/(n+mu))
    d = 1+racine(mu/n)
    d_i = 3*n
    x=np.array(x)
    
   # x = 1000 *np.random.normal(0, 1, n)  # on prend juste un individu appartenant à R2 pour initialiser 
    sigma = np.abs(np.random.normal(0, 1, n))
    s_sigma=np.zeros((n,1))
    
    
   
        
    for e in range(epochs): 
        
        liste_xk =[]
        liste_zk =[]
        for k in range(lambda_):
        
            z_k = np.random.normal(0, 1, n)
            liste_zk.append(z_k)
            x_k= x + 0.1*z_k  #x_k= x + sigma*z_k     RESOUDRE PROBLEME AVEC SIGMA 
            liste_xk.append(x_k)
   
        parent = select_mu_best(liste_xk, mu, n, func)
        

        s_sigma =  s_sigma*(1- c_sigma)+ np.sum(np.array(liste_zk))*racine((2-c_sigma)*c_sigma)*racine(mu)/mu
        #val_normale = np.abs(np.random.normal(0, 1, 1))
        val_normale = np.sqrt(2/math.pi)
        vals_normale = n
        #val_normale = np.abs(np.random.normal(0, 1, 1))
        #vals_normale = np.abs(np.random.normal(0, 1, n)) # 2 espace de dimension à optim
        # sigma= sigma* np.exp((1/d_i)* ((np.abs(s_sigma[0])/val_normale)-1))*np.exp((c_sigma/d)* ((norm_2(s_sigma)/norm_2(vals_normale))-1))

        
        sigma= sigma* np.exp((1/d_i)* ((np.abs(s_sigma[0])/val_normale)-1))*np.exp((c_sigma/d)* ((norm_2(s_sigma)/n)-1))
        x = np.mean(np.array(parent), axis= 0)
        print("resultat f(x): ",fitness(func , x), "avec X = ", x)
    return x 
            
      
        
        
   
            
        
    
    
        
        
        
        
    
```
